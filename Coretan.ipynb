```python
import pandas as pd
import matplotlib.pyplot as plt
import streamlit as st
import seaborn as sns
import openpyxl
```

```python
df_listrik = pd.read_excel("D:\TetrisProgram\Capstone Project\harga_listrik.xlsx")
df_listrik['tahun'] = df_listrik.loc[:,'tahun'].astype('str')
df_listrik.head()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_listrik, x='tahun', y='harga', hue='wilayah')
plt.title('Nilai Rp/KWh Listrik di Jawa Tengah menurut unit PLN', fontsize=15, loc='center', pad=15)
plt.ylabel('Harga listrik (Rp/KWh)')
plt.grid(color='gray', linewidth=0.5, linestyle=':')
plt.ylim(ymin=0)
plt.tight_layout()
plt.show()
```

```python
df_bahanbakar = pd.read_excel('D:\TetrisProgram\Capstone Project\\bahanbakar_pembangkit_listrik.xlsx')
df_bahanbakar
```

```python
fig, ax = plt.subplots()
fig.set_facecolor('white')
ax.set_title('Komposisi Penggunaan Bahan Bakar Pembangkit Listrik\nTahun 2020 di Indonesia', fontsize=15)
patches, texts, autotexts = ax.pie(df_bahanbakar['persentase'], labels=df_bahanbakar['Bahan Bakar'], autopct='%.0f%%', textprops={'fontsize': 15})
for i in texts:
    i.set_fontsize(12)
plt.tight_layout()
plt.show()
```

```python
df_coal = pd.read_excel('D:\TetrisProgram\Capstone Project\harga_batubara.xlsx')
df_coal['tahun'] = df_coal['tahun'].astype('str')
df_coal_mean = df_coal.groupby('tahun')['harga'].mean().reset_index()
df_coal_mean
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_coal_mean, x='tahun', y='harga')
plt.title('Harga Rata-Rata Batubara per Ton(dalam USD)', fontsize=15, pad=10, loc='center')
plt.ylabel('Harga batubara (USD/ton)')
plt.grid(color='gray', linewidth=0.5, linestyle=':')
plt.ylim(ymin=0)
plt.tight_layout()
plt.show()
```

```python
df_pengguna = pd.read_excel("D:\TetrisProgram\Capstone Project\jumlah_pengguna.xlsx")
df_pengguna['tahun'] = df_pengguna['tahun'].astype('str')
df_pengguna['jumlah'] = df_pengguna['jumlah']/1000000
df_pengguna.head()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df_pengguna, x='tahun', y='jumlah')
plt.title('Jumlah Listrik yang Didistribusikan di Indonesia\n Periode 2012-2020', fontsize=15)
plt.ylabel('jumlah energi yang terjual(juta KWh)')
plt.grid(linewidth=0.5, linestyle=":", color='darkgray')
plt.tight_layout()
plt.show()
```
